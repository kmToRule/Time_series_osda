```python
#Baseline model
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from torch.utils.data import TensorDataset, DataLoader

# Load and preprocess WISDM dataset
def load_data(file_path):
    data = []
    with open(file_path) as f:
        for line in f:
            parts = line.strip().split(';')[0].split(',')
            if len(parts) != 6:
                continue
            try:
                user, activity, timestamp, x, y, z = parts
                data.append([
                    int(user),
                    activity.strip(),
                    int(timestamp) // 10**9,
                    float(x), float(y), float(z)
                ])
            except:
                continue
    df = pd.DataFrame(data, columns=['user', 'activity', 'timestamp', 'x', 'y', 'z'])
    for col in ['x', 'y', 'z']:
        df[col] = (df[col] - df[col].mean()) / df[col].std()
    return df

df = load_data('WISDM.txt')

known_classes = ['Walking', 'Jogging', 'Sitting']
unknown_classes = ['Standing', 'Upstairs', 'Downstairs']

le = LabelEncoder()
le.fit(known_classes)

df_source = df[df['activity'].isin(known_classes)]
df_target = df[df['activity'].isin(known_classes + unknown_classes)].copy()
df_target['open_label'] = df_target['activity'].apply(lambda x: le.transform([x])[0] if x in known_classes else -1)


def segment_data(df, label_col='activity', window_size=60, step_size=20):
    segments, labels = [], []
    for user in df['user'].unique():
        user_data = df[df['user'] == user]
        for activity in user_data[label_col].unique():
            activity_data = user_data[user_data[label_col] == activity]
            if len(activity_data) < window_size:
                continue
            for start in range(0, len(activity_data) - window_size + 1, step_size):
                end = start + window_size
                segment = activity_data.iloc[start:end]
                if len(segment) == window_size:
                    x = segment['x'].values
                    y = segment['y'].values
                    z = segment['z'].values
                    segments.append(np.vstack([x, y, z]).T)
                    labels.append(activity)
    return np.array(segments), np.array(labels)

X_s, y_s = segment_data(df_source)
X_t, y_t = segment_data(df_target, label_col='open_label')

y_s_encoded = le.transform(y_s)

X_s_train, X_s_test, y_s_train, y_s_test = train_test_split(
    X_s, y_s_encoded, test_size=0.2, random_state=42, stratify=y_s_encoded)

class FrequencyEncoder(nn.Module):
    def __init__(self, input_size, hidden_size):
        super().__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.bn1 = nn.BatchNorm1d(hidden_size)

    def forward(self, x):
        x_freq = torch.fft.fft(x, dim=1)
        x_mag = torch.abs(x_freq).real
        x_flat = x_mag.reshape(x_mag.size(0), -1)
        return F.relu(self.bn1(self.fc1(x_flat)))

class OpenSetClassifier(nn.Module):
    def __init__(self, input_size, num_classes):
        super().__init__()
        self.fc = nn.Linear(input_size, num_classes)

    def forward(self, x):
        logits = self.fc(x)
        max_prob = torch.max(torch.softmax(logits, dim=1), dim=1)[0]
        return logits, max_prob

class RAINCOAT(nn.Module):
    def __init__(self, encoder, classifier):
        super().__init__()
        self.encoder = encoder
        self.classifier = classifier

    def forward(self, x):
        features = self.encoder(x)
        logits, max_prob = self.classifier(features)
        return logits, max_prob, features

def mmd_loss(source, target):
    def gaussian_kernel(x, y, sigma=1.0):
        x = x.unsqueeze(1)
        y = y.unsqueeze(0)
        return torch.exp(-((x - y)**2).sum(2) / (2 * sigma**2))
    K_ss = gaussian_kernel(source, source)
    K_tt = gaussian_kernel(target, target)
    K_st = gaussian_kernel(source, target)
    return K_ss.mean() + K_tt.mean() - 2 * K_st.mean()

def train(model, loader_s, loader_t, optimizer, criterion, device, threshold=0.99):
    model.train()
    for epoch in range(10):
        for (x_s, y_s), (x_t, _) in zip(loader_s, loader_t):
            x_s, y_s = x_s.to(device), y_s.to(device)
            x_t = x_t.to(device)

            logits_s, _, f_s = model(x_s)
            _, _, f_t = model(x_t)

            loss_cls = criterion(logits_s, y_s)
            loss_mmd = mmd_loss(f_s, f_t)

            loss = loss_cls + 0.1 * loss_mmd
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
        print(f"Epoch {epoch+1}: Loss = {loss.item():.4f}")

def evaluate(model, X, y, device, threshold=0.99):
    model.eval()
    y_pred, y_true = [], []
    with torch.no_grad():
        for i in range(0, len(X), 64):
            x_batch = torch.tensor(X[i:i+64], dtype=torch.float32).to(device)
            y_batch = y[i:i+64]
            logits, max_prob, _ = model(x_batch)
            preds = torch.argmax(logits, dim=1).cpu().numpy()
            max_p = max_prob.cpu().numpy()
            preds[max_p < threshold] = -1
            y_pred.extend(preds)
            y_true.extend(y_batch)

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    known_mask = y_true != -1
    unknown_mask = y_true == -1

    acc_known = accuracy_score(y_true[known_mask], y_pred[known_mask])
    acc_unknown = accuracy_score(y_true[unknown_mask], y_pred[unknown_mask])

    if (acc_known + acc_unknown) == 0:
        h_score = 0.0
    else:
        h_score = 2 * acc_known * acc_unknown / (acc_known + acc_unknown)

    return acc_known, acc_unknown, h_score

# Run all
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

X_s_train_tensor = torch.tensor(X_s_train, dtype=torch.float32)
y_s_train_tensor = torch.tensor(y_s_train, dtype=torch.long)
X_t_tensor = torch.tensor(X_t, dtype=torch.float32)
y_t_tensor = torch.tensor(y_t, dtype=torch.long)

train_loader_s = DataLoader(TensorDataset(X_s_train_tensor, y_s_train_tensor), batch_size=64, shuffle=True)
train_loader_t = DataLoader(TensorDataset(X_t_tensor, y_t_tensor), batch_size=64, shuffle=True)

input_size = 60 * 3
encoder = FrequencyEncoder(input_size=input_size, hidden_size=128)
classifier = OpenSetClassifier(input_size=128, num_classes=len(known_classes))
model = RAINCOAT(encoder, classifier).to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.CrossEntropyLoss()

train(model, train_loader_s, train_loader_t, optimizer, criterion, device)

acc_known, acc_unknown, h_score = evaluate(model, X_t, y_t, device)
print(f"Known Accuracy: {acc_known:.4f}")
print(f"Unknown Accuracy: {acc_unknown:.4f}")
print(f"H-Score: {h_score:.4f}")
```

```python
# Full Integrated CALOT-based OSDA for WISDM Time-Series Data

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, pairwise_distances
from torch.utils.data import TensorDataset, DataLoader

# ---------------------------------------------
# Data Loading and Preprocessing
# ---------------------------------------------

def load_data(file_path):
    data = []
    with open(file_path) as f:
        for line in f:
            parts = line.strip().split(';')[0].split(',')
            if len(parts) != 6:
                continue
            try:
                user, activity, timestamp, x, y, z = parts
                data.append([
                    int(user),
                    activity.strip(),
                    int(timestamp) // 10**9,
                    float(x), float(y), float(z)
                ])
            except:
                continue
    df = pd.DataFrame(data, columns=['user', 'activity', 'timestamp', 'x', 'y', 'z'])
    for col in ['x', 'y', 'z']:
        df[col] = (df[col] - df[col].mean()) / df[col].std()
    return df

# ---------------------------------------------
# Segmentation
# ---------------------------------------------

def segment_data(df, label_col='activity', window_size=60, step_size=20):
    segments, labels = [], []
    for user in df['user'].unique():
        user_data = df[df['user'] == user]
        for activity in user_data[label_col].unique():
            activity_data = user_data[user_data[label_col] == activity]
            if len(activity_data) < window_size:
                continue
            for start in range(0, len(activity_data) - window_size + 1, step_size):
                end = start + window_size
                segment = activity_data.iloc[start:end]
                if len(segment) == window_size:
                    x = segment['x'].values
                    y = segment['y'].values
                    z = segment['z'].values
                    segments.append(np.vstack([x, y, z]).T)
                    labels.append(activity)
    return np.array(segments), np.array(labels)

# ---------------------------------------------
# Models
# ---------------------------------------------

class FrequencyEncoder(nn.Module):
    def __init__(self, input_size, hidden_size):
        super().__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.bn1 = nn.BatchNorm1d(hidden_size)

    def forward(self, x):
        x_freq = torch.fft.fft(x, dim=1)
        x_mag = torch.abs(x_freq).real
        x_flat = x_mag.reshape(x_mag.size(0), -1)
        return F.relu(self.bn1(self.fc1(x_flat)))

class OpenSetClassifier(nn.Module):
    def __init__(self, input_size, num_classes):
        super().__init__()
        self.fc = nn.Linear(input_size, num_classes)

    def forward(self, x):
        logits = self.fc(x)
        max_prob = torch.max(torch.softmax(logits, dim=1), dim=1)[0]
        return logits, max_prob

class RAINCOAT(nn.Module):
    def __init__(self, encoder, classifier):
        super().__init__()
        self.encoder = encoder
        self.classifier = classifier

    def forward(self, x):
        features = self.encoder(x)
        logits, max_prob = self.classifier(features)
        return logits, max_prob, features

# ---------------------------------------------
# CALOT Utilities
# ---------------------------------------------

def compute_confidence(logits):
    probs = F.softmax(logits, dim=-1)
    return torch.max(probs, dim=-1)[0]

def entropy_of_prediction(logits):
    probs = F.softmax(logits, dim=-1)
    entropy = -torch.sum(probs * torch.log(probs + 1e-8), dim=-1)
    return entropy

def latent_confidence_based_mixing(source_feats, target_feats, target_logits, threshold=0.7):
    confidence = compute_confidence(target_logits)
    mask = (confidence > threshold).float().unsqueeze(-1)
    mixed = mask * target_feats + (1 - mask) * source_feats
    return mixed, confidence

def hierarchical_ot(source_feats, target_feats):
    B, D = source_feats.shape
    costs = torch.zeros(B)
    for i in range(B):
        src = source_feats[i].view(-1, 1).detach().cpu().numpy()
        tgt = target_feats[i].view(-1, 1).detach().cpu().numpy()
        C = pairwise_distances(src, tgt, metric='euclidean')
        costs[i] = torch.tensor(C.mean())
    return costs


def open_set_regularization(ot_costs, margin=1.0):
    penalties = torch.clamp(ot_costs - margin, min=0.0)
    return penalties.mean()

def entropy_filtered_transport_map(ot_costs, logits):
    entropy = entropy_of_prediction(logits)  # shape: (B,)
    weights = torch.exp(-entropy).to(ot_costs.device)  # Ensure same device
    return (weights * ot_costs).mean()



def calot_step(source_feats, target_feats, target_logits, margin=1.0, conf_threshold=0.7):
    mixed_feats, confidence = latent_confidence_based_mixing(source_feats, target_feats, target_logits, threshold=conf_threshold)
    ot_costs = hierarchical_ot(source_feats, mixed_feats)
    reg_loss = open_set_regularization(ot_costs, margin=margin)
    transport_loss = entropy_filtered_transport_map(ot_costs, target_logits)
    return reg_loss + transport_loss, reg_loss, transport_loss

# ---------------------------------------------
# Training & Evaluation
# ---------------------------------------------

def train_calot(model, loader_s, loader_t, optimizer, criterion, device):
    model.train()
    for epoch in range(10):
        for (x_s, y_s), (x_t, _) in zip(loader_s, loader_t):
          x_s, y_s = x_s.to(device), y_s.to(device)
          x_t = x_t.to(device)

          # Align batch sizes
          min_len = min(x_s.size(0), x_t.size(0))
          x_s, y_s = x_s[:min_len], y_s[:min_len]
          x_t = x_t[:min_len]

          logits_s, _, f_s = model(x_s)
          logits_t, _, f_t = model(x_t)

          loss_cls = criterion(logits_s, y_s)
          calot_loss, reg, transp = calot_step(f_s, f_t, logits_t)

          loss = loss_cls + 0.1 * calot_loss

          optimizer.zero_grad()
          loss.backward()
          optimizer.step()

        print(f"Epoch {epoch+1}: Loss = {loss.item():.4f} | CALOT Loss = {calot_loss.item():.4f}")

def evaluate(model, X, y, device, threshold=0.99):
    model.eval()
    y_pred, y_true = [], []
    with torch.no_grad():
        for i in range(0, len(X), 64):
            x_batch = torch.tensor(X[i:i+64], dtype=torch.float32).to(device)
            y_batch = y[i:i+64]
            logits, max_prob, _ = model(x_batch)
            preds = torch.argmax(logits, dim=1).cpu().numpy()
            max_p = max_prob.cpu().numpy()
            preds[max_p < threshold] = -1
            y_pred.extend(preds)
            y_true.extend(y_batch)

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    known_mask = y_true != -1
    unknown_mask = y_true == -1
    acc_known = accuracy_score(y_true[known_mask], y_pred[known_mask])
    acc_unknown = accuracy_score(y_true[unknown_mask], y_pred[unknown_mask])
    h_score = (2 * acc_known * acc_unknown / (acc_known + acc_unknown)) if (acc_known + acc_unknown) else 0.0
    return acc_known, acc_unknown, h_score

# ---------------------------------------------
# Main Execution
# ---------------------------------------------

file_path = '/content/drive/My Drive/Raincoat/data/WISDM.txt'
df = load_data(file_path)
print("Unique Activity Labels in WISDM dataset:")
print(df['activity'].dropna().unique())
known_classes = ['Walking', 'Jogging', 'Sitting','Standing']
unknown_classes = ['Upstairs', 'Downstairs']
le = LabelEncoder()
le.fit(known_classes)

df_source = df[df['activity'].isin(known_classes)]
df_target = df[df['activity'].isin(known_classes + unknown_classes)].copy()
df_target['open_label'] = df_target['activity'].apply(lambda x: le.transform([x])[0] if x in known_classes else -1)

X_s, y_s = segment_data(df_source)
X_t, y_t = segment_data(df_target, label_col='open_label')
y_s_encoded = le.transform(y_s)
X_s_train, X_s_test, y_s_train, y_s_test = train_test_split(X_s, y_s_encoded, test_size=0.2, random_state=42, stratify=y_s_encoded)

# Dataloaders
X_s_train_tensor = torch.tensor(X_s_train, dtype=torch.float32)
y_s_train_tensor = torch.tensor(y_s_train, dtype=torch.long)
X_t_tensor = torch.tensor(X_t, dtype=torch.float32)
y_t_tensor = torch.tensor(y_t, dtype=torch.long)
train_loader_s = DataLoader(TensorDataset(X_s_train_tensor, y_s_train_tensor), batch_size=64, shuffle=True)
train_loader_t = DataLoader(TensorDataset(X_t_tensor, y_t_tensor), batch_size=64, shuffle=True)

# Model Init
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
input_size = 60 * 3
encoder = FrequencyEncoder(input_size=input_size, hidden_size=128)
classifier = OpenSetClassifier(input_size=128, num_classes=len(known_classes))
model = RAINCOAT(encoder, classifier).to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.CrossEntropyLoss()

train_calot(model, train_loader_s, train_loader_t, optimizer, criterion, device)
acc_known, acc_unknown, h_score = evaluate(model, X_t, y_t, device)
print(f"Known Accuracy: {acc_known:.4f}")
print(f"Unknown Accuracy: {acc_unknown:.4f}")
print(f"H-Score: {h_score:.4f}")
# ---------------------------------------------
# t-SNE Visualization (Before vs After CALOT)
# ---------------------------------------------
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import os

def extract_features(model, X, device):
    model.eval()
    feats, labels = [], []
    with torch.no_grad():
        for i in range(0, len(X), 64):
            x_batch = torch.tensor(X[i:i+64], dtype=torch.float32).to(device)
            _, _, f = model(x_batch)
            feats.append(f.cpu().numpy())
    return np.vstack(feats)

# 1. Extract source features before CALOT (from untrained model)
encoder_pre = FrequencyEncoder(input_size=input_size, hidden_size=128).to(device)
classifier_pre = OpenSetClassifier(input_size=128, num_classes=len(known_classes)).to(device)
model_pre = RAINCOAT(encoder_pre, classifier_pre).to(device)
feats_before = extract_features(model_pre, X_t, device)

# 2. Extract source features after CALOT (trained model)
feats_after = extract_features(model, X_t, device)

# 3. t-SNE reduction
tsne = TSNE(n_components=2, random_state=42)
reduced_before = tsne.fit_transform(feats_before)
reduced_after = tsne.fit_transform(feats_after)

# 4. Labels for coloring
labels_t = np.array([str(l) if l != -1 else 'Unknown' for l in y_t])

# 5. Save to Drive
output_dir = '/content/drive/MyDrive/WISDM_TSNE_Plots'
os.makedirs(output_dir, exist_ok=True)
save_path = os.path.join(output_dir, 'tsne_before_vs_after_calot.png')

def plot_tsne_side_by_side(feats_before, labels_before, feats_after, labels_after, save_path):
    fig, axes = plt.subplots(1, 2, figsize=(16, 7))
    for ax, feats, labels, title in zip(
        axes,
        [feats_before, feats_after],
        [labels_before, labels_after],
        ["Before CALOT Adaptation", "After CALOT Adaptation"]
    ):
        unique = np.unique(labels)
        for label in unique:
            idx = np.where(np.array(labels) == label)
            ax.scatter(feats[idx, 0], feats[idx, 1], label=label, alpha=0.6, s=30)
        ax.set_title(title)
        ax.set_xlabel("t-SNE Dim 1")
        ax.set_ylabel("t-SNE Dim 2")

    handles, labels = axes[1].get_legend_handles_labels()
    fig.legend(handles, labels, loc='center left', bbox_to_anchor=(1.02, 0.5), fontsize='small')
    plt.tight_layout()
    plt.savefig(save_path)
    print(f"Saved t-SNE subplot to: {save_path}")
    plt.show()

plot_tsne_side_by_side(reduced_before, labels_t, reduced_after, labels_t, save_path)

```

```python
# ABLATION-1

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, pairwise_distances
from torch.utils.data import TensorDataset, DataLoader

# ---------------------------------------------
# Data Loading and Preprocessing
# ---------------------------------------------

def load_data(file_path):
    data = []
    with open(file_path) as f:
        for line in f:
            parts = line.strip().split(';')[0].split(',')
            if len(parts) != 6:
                continue
            try:
                user, activity, timestamp, x, y, z = parts
                data.append([
                    int(user),
                    activity.strip(),
                    int(timestamp) // 10**9,
                    float(x), float(y), float(z)
                ])
            except:
                continue
    df = pd.DataFrame(data, columns=['user', 'activity', 'timestamp', 'x', 'y', 'z'])
    for col in ['x', 'y', 'z']:
        df[col] = (df[col] - df[col].mean()) / df[col].std()
    return df

# ---------------------------------------------
# Segmentation
# ---------------------------------------------

def segment_data(df, label_col='activity', window_size=60, step_size=20):
    segments, labels = [], []
    for user in df['user'].unique():
        user_data = df[df['user'] == user]
        for activity in user_data[label_col].unique():
            activity_data = user_data[user_data[label_col] == activity]
            if len(activity_data) < window_size:
                continue
            for start in range(0, len(activity_data) - window_size + 1, step_size):
                end = start + window_size
                segment = activity_data.iloc[start:end]
                if len(segment) == window_size:
                    x = segment['x'].values
                    y = segment['y'].values
                    z = segment['z'].values
                    segments.append(np.vstack([x, y, z]).T)
                    labels.append(activity)
    return np.array(segments), np.array(labels)

# ---------------------------------------------
# Models
# ---------------------------------------------

class FrequencyEncoder(nn.Module):
    def __init__(self, input_size, hidden_size):
        super().__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.bn1 = nn.BatchNorm1d(hidden_size)

    def forward(self, x):
        x_freq = torch.fft.fft(x, dim=1)
        x_mag = torch.abs(x_freq).real
        x_flat = x_mag.reshape(x_mag.size(0), -1)
        return F.relu(self.bn1(self.fc1(x_flat)))

class OpenSetClassifier(nn.Module):
    def __init__(self, input_size, num_classes):
        super().__init__()
        self.fc = nn.Linear(input_size, num_classes)

    def forward(self, x):
        logits = self.fc(x)
        max_prob = torch.max(torch.softmax(logits, dim=1), dim=1)[0]
        return logits, max_prob

class RAINCOAT(nn.Module):
    def __init__(self, encoder, classifier):
        super().__init__()
        self.encoder = encoder
        self.classifier = classifier

    def forward(self, x):
        features = self.encoder(x)
        logits, max_prob = self.classifier(features)
        return logits, max_prob, features

# ---------------------------------------------
# CALOT Utilities
# ---------------------------------------------

def compute_confidence(logits):
    probs = F.softmax(logits, dim=-1)
    return torch.max(probs, dim=-1)[0]

def entropy_of_prediction(logits):
    probs = F.softmax(logits, dim=-1)
    entropy = -torch.sum(probs * torch.log(probs + 1e-8), dim=-1)
    return entropy

def latent_confidence_based_mixing(source_feats, target_feats, target_logits, threshold=0.7):
    confidence = compute_confidence(target_logits)
    mask = (confidence > threshold).float().unsqueeze(-1)
    mixed = mask * target_feats + (1 - mask) * source_feats
    return mixed, confidence

def hierarchical_ot(source_feats, target_feats):
    B, D = source_feats.shape
    costs = torch.zeros(B)
    for i in range(B):
        src = source_feats[i].view(-1, 1).detach().cpu().numpy()
        tgt = target_feats[i].view(-1, 1).detach().cpu().numpy()
        C = pairwise_distances(src, tgt, metric='euclidean')
        costs[i] = torch.tensor(C.mean())
    return costs


def open_set_regularization(ot_costs, margin=1.0):
    penalties = torch.clamp(ot_costs - margin, min=0.0)
    return penalties.mean()

def entropy_filtered_transport_map(ot_costs, logits):
    entropy = entropy_of_prediction(logits)  # shape: (B,)
    weights = torch.exp(-entropy).to(ot_costs.device)  # Ensure same device
    return (weights * ot_costs).mean()


def calot_step_no_confidence_mask(source_feats, target_feats, target_logits, margin=1.0):
    # Skip masking: no confidence threshold
    mixed_feats = target_feats  # just use target features directly
    ot_costs = hierarchical_ot(source_feats, mixed_feats)
    reg_loss = open_set_regularization(ot_costs, margin=margin)
    transport_loss = entropy_filtered_transport_map(ot_costs, target_logits)
    return reg_loss + transport_loss, reg_loss, transport_loss

# ---------------------------------------------
# Training & Evaluation
# ---------------------------------------------

def train_calot(model, loader_s, loader_t, optimizer, criterion, device):
    model.train()
    for epoch in range(10):
        for (x_s, y_s), (x_t, _) in zip(loader_s, loader_t):
          x_s, y_s = x_s.to(device), y_s.to(device)
          x_t = x_t.to(device)

          # Align batch sizes
          min_len = min(x_s.size(0), x_t.size(0))
          x_s, y_s = x_s[:min_len], y_s[:min_len]
          x_t = x_t[:min_len]

          logits_s, _, f_s = model(x_s)
          logits_t, _, f_t = model(x_t)

          loss_cls = criterion(logits_s, y_s)
          calot_loss, reg, transp = calot_step_no_confidence_mask(f_s, f_t, logits_t)

          loss = loss_cls + 0.1 * calot_loss

          optimizer.zero_grad()
          loss.backward()
          optimizer.step()

        print(f"Epoch {epoch+1}: Loss = {loss.item():.4f} | CALOT Loss = {calot_loss.item():.4f}")

def evaluate(model, X, y, device, threshold=0.99):
    model.eval()
    y_pred, y_true = [], []
    with torch.no_grad():
        for i in range(0, len(X), 64):
            x_batch = torch.tensor(X[i:i+64], dtype=torch.float32).to(device)
            y_batch = y[i:i+64]
            logits, max_prob, _ = model(x_batch)
            preds = torch.argmax(logits, dim=1).cpu().numpy()
            max_p = max_prob.cpu().numpy()
            preds[max_p < threshold] = -1
            y_pred.extend(preds)
            y_true.extend(y_batch)

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    known_mask = y_true != -1
    unknown_mask = y_true == -1
    acc_known = accuracy_score(y_true[known_mask], y_pred[known_mask])
    acc_unknown = accuracy_score(y_true[unknown_mask], y_pred[unknown_mask])
    h_score = (2 * acc_known * acc_unknown / (acc_known + acc_unknown)) if (acc_known + acc_unknown) else 0.0
    return acc_known, acc_unknown, h_score

# ---------------------------------------------
# Main Execution
# ---------------------------------------------

file_path = '/content/drive/My Drive/Raincoat/data/WISDM.txt'
df = load_data(file_path)
known_classes = ['Walking', 'Jogging', 'Sitting','Standing']
unknown_classes = ['Upstairs', 'Downstairs']
le = LabelEncoder()
le.fit(known_classes)

df_source = df[df['activity'].isin(known_classes)]
df_target = df[df['activity'].isin(known_classes + unknown_classes)].copy()
df_target['open_label'] = df_target['activity'].apply(lambda x: le.transform([x])[0] if x in known_classes else -1)

X_s, y_s = segment_data(df_source)
X_t, y_t = segment_data(df_target, label_col='open_label')
y_s_encoded = le.transform(y_s)
X_s_train, X_s_test, y_s_train, y_s_test = train_test_split(X_s, y_s_encoded, test_size=0.2, random_state=42, stratify=y_s_encoded)

# Dataloaders
X_s_train_tensor = torch.tensor(X_s_train, dtype=torch.float32)
y_s_train_tensor = torch.tensor(y_s_train, dtype=torch.long)
X_t_tensor = torch.tensor(X_t, dtype=torch.float32)
y_t_tensor = torch.tensor(y_t, dtype=torch.long)
train_loader_s = DataLoader(TensorDataset(X_s_train_tensor, y_s_train_tensor), batch_size=64, shuffle=True)
train_loader_t = DataLoader(TensorDataset(X_t_tensor, y_t_tensor), batch_size=64, shuffle=True)

# Model Init
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
input_size = 60 * 3
encoder = FrequencyEncoder(input_size=input_size, hidden_size=128)
classifier = OpenSetClassifier(input_size=128, num_classes=len(known_classes))
model = RAINCOAT(encoder, classifier).to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.CrossEntropyLoss()

train_calot(model, train_loader_s, train_loader_t, optimizer, criterion, device)
acc_known, acc_unknown, h_score = evaluate(model, X_t, y_t, device)
print(f"Known Accuracy: {acc_known:.4f}")
print(f"Unknown Accuracy: {acc_unknown:.4f}")
print(f"H-Score: {h_score:.4f}")
```

```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import accuracy_score

from sklearn.metrics import pairwise_distances
""
# For better CUDA error messages
os.environ["CUDA_LAUNCH_BLOCKING"] = "1"
torch.autograd.set_detect_anomaly(True)

# -------------------------------
# HHAR_SA Loader
# -------------------------------

def load_hhar_data(data_dir, domain_ids, known_classes):
    X_s, y_s, X_t, y_t = [], [], [], []

    for domain_id in domain_ids:
        train_path = f"{data_dir}/train_{domain_id}.pt"
        test_path = f"{data_dir}/test_{domain_id}.pt"

        train_data = torch.load(train_path, weights_only=False)
        test_data = torch.load(test_path, weights_only=False)

        # Safely extract arrays
        X_train = train_data['samples']
        y_train = train_data['labels']
        X_test = test_data['samples']
        y_test = test_data['labels']

        # Ensure they are numpy arrays
        if isinstance(X_train, torch.Tensor): X_train = X_train.numpy()
        if isinstance(y_train, torch.Tensor): y_train = y_train.numpy()
        if isinstance(X_test, torch.Tensor): X_test = X_test.numpy()
        if isinstance(y_test, torch.Tensor): y_test = y_test.numpy()

        # Normalize
        mean = X_train.mean()
        std = X_train.std()
        X_train = (X_train - mean) / std
        X_test = (X_test - mean) / std

        # Known class filtering for source domain
        source_mask = np.isin(y_train, known_classes)
        X_s.append(torch.tensor(X_train[source_mask], dtype=torch.float32))
        y_s.append(torch.tensor(y_train[source_mask], dtype=torch.long))

        # Target domain: relabel unknowns as -1
        target_labels = [label if label in known_classes else -1 for label in y_test]
        X_t.append(torch.tensor(X_test, dtype=torch.float32))
        y_t.append(torch.tensor(target_labels, dtype=torch.long))

    return torch.cat(X_s), torch.cat(y_s), torch.cat(X_t), torch.cat(y_t)


# Model Definitions
# -------------------------------

class FrequencyEncoder(nn.Module):
    def __init__(self, input_size, hidden_size):
        super().__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.bn1 = nn.BatchNorm1d(hidden_size)

    def forward(self, x):
        # x: [B, 128, 3]  FFT on time axis (dim=1)
        x_freq = torch.fft.fft(x, dim=1)  # [B, 128, 3], complex64
        x_mag = torch.abs(x_freq)        # take magnitude
        x_flat = x_mag.reshape(x.size(0), -1).contiguous()  # [B, 384]
        return F.relu(self.bn1(self.fc1(x_flat)))


class OpenSetClassifier(nn.Module):
    def __init__(self, input_size, num_classes):
        super().__init__()
        self.fc = nn.Linear(input_size, num_classes)

    def forward(self, x):
        logits = self.fc(x)
        max_prob = torch.max(torch.softmax(logits, dim=1), dim=1)[0]
        return logits, max_prob

class RAINCOAT(nn.Module):
    def __init__(self, encoder, classifier):
        super().__init__()
        self.encoder = encoder
        self.classifier = classifier

    def forward(self, x):
        features = self.encoder(x)
        logits, max_prob = self.classifier(features)
        return logits, max_prob, features


# -------------------------------
# CALOT Utilities (unchanged)
# -------------------------------

def compute_confidence(logits):
    probs = F.softmax(logits, dim=-1)
    return torch.max(probs, dim=-1)[0]

def entropy_of_prediction(logits):
    probs = F.softmax(logits, dim=-1)
    entropy = -torch.sum(probs * torch.log(probs + 1e-8), dim=-1)
    return entropy

def latent_confidence_based_mixing(source_feats, target_feats, target_logits, threshold=0.7):
    confidence = compute_confidence(target_logits)
    mask = (confidence > threshold).float().unsqueeze(-1)
    mixed = mask * target_feats + (1 - mask) * source_feats
    return mixed, confidence

def hierarchical_ot(source_feats, target_feats):
    B, D = source_feats.shape
    costs = torch.zeros(B)
    for i in range(B):
        src = source_feats[i].view(-1, 1).detach().cpu().numpy()
        tgt = target_feats[i].view(-1, 1).detach().cpu().numpy()
        C = pairwise_distances(src, tgt, metric='euclidean')
        costs[i] = torch.tensor(C.mean())
    return costs

def open_set_regularization(ot_costs, margin=1.0):
    penalties = torch.clamp(ot_costs - margin, min=0.0)
    return penalties.mean()

def entropy_filtered_transport_map(ot_costs, logits):
    entropy = entropy_of_prediction(logits)
    weights = torch.exp(-entropy).to(ot_costs.device)
    return (weights * ot_costs).mean()

def calot_step(source_feats, target_feats, target_logits, margin=1.0, conf_threshold=0.7):
    mixed_feats, confidence = latent_confidence_based_mixing(source_feats, target_feats, target_logits, threshold=conf_threshold)
    ot_costs = hierarchical_ot(source_feats, mixed_feats)
    reg_loss = open_set_regularization(ot_costs, margin=margin)
    transport_loss = entropy_filtered_transport_map(ot_costs, target_logits)
    return reg_loss + transport_loss, reg_loss, transport_loss


# -------------------------------
# Training and Evaluation
# -------------------------------

def train_calot(model, loader_s, loader_t, optimizer, criterion, device):
    model.train()
    for epoch in range(10):
        for (x_s, y_s), (x_t,) in zip(loader_s, loader_t):
            x_s, y_s = x_s.to(device), y_s.to(device)
            x_t = x_t.to(device)

            min_len = min(x_s.size(0), x_t.size(0))
            x_s, y_s, x_t = x_s[:min_len], y_s[:min_len], x_t[:min_len]
            x_s = x_s.view(-1, 128, 3).to(device)
            x_t = x_t.view(-1, 128, 3).to(device)
            logits_s, _, f_s = model(x_s)
            logits_t, _, f_t = model(x_t)

            # Skip unknown labels (-1) for loss computation
            mask = y_s != -1
            if mask.sum() == 0:
                continue

            inputs = x_s[mask]
            labels = y_s[mask]

            loss_cls = criterion(logits_s[mask], labels)
            calot_loss, reg, transp = calot_step(f_s, f_t, logits_t)

            loss = loss_cls + 0.1 * calot_loss

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        print(f"Epoch {epoch+1}: Loss = {loss.item():.4f} | CALOT Loss = {calot_loss.item():.4f}")

def evaluate(model, X, y, device, threshold=0.95):
    model.eval()
    y_pred, y_true = [], []
    with torch.no_grad():
        for i in range(0, len(X), 64):
            x_batch = torch.tensor(X[i:i+64], dtype=torch.float32).to(device)
            x_batch = x_batch.view(-1, 128, 3).to(device)
            y_batch = y[i:i+64]

            logits, max_prob, _ = model(x_batch)
            preds = torch.argmax(logits, dim=1).cpu().numpy()
            max_p = max_prob.cpu().numpy()
            preds[max_p < threshold] = -1
            y_pred.extend(preds)
            y_true.extend(y_batch)

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    known_mask = y_true != -1
    unknown_mask = y_true == -1
    acc_known = accuracy_score(y_true[known_mask], y_pred[known_mask])
    acc_unknown = accuracy_score(y_true[unknown_mask], y_pred[unknown_mask])
    h_score = (2 * acc_known * acc_unknown / (acc_known + acc_unknown)) if (acc_known + acc_unknown) else 0.0
    return acc_known, acc_unknown, h_score

# -------------------------------
# Main Execution
# -------------------------------

known_classes = [0, 1, 2, 3]
unknown_classes = [4, 5]

data_path = '/content/drive/My Drive/Raincoat/data/HHAR_SA'
X_s, y_s, X_t, y_t = load_hhar_data(data_path, domain_ids=[0], known_classes=known_classes)

# Split source for train/test
X_s_train, X_s_test, y_s_train, y_s_test = train_test_split(
    X_s.numpy(), y_s.numpy(), test_size=0.2, stratify=y_s.numpy(), random_state=42
)

X_s_train = torch.tensor(X_s_train, dtype=torch.float32)
y_s_train = torch.tensor(y_s_train, dtype=torch.long)

# Only use X_t for target (no y_t during training!)
X_t_tensor = torch.tensor(X_t, dtype=torch.float32)

# Dataloaders
train_loader_s = DataLoader(TensorDataset(X_s_train, y_s_train), batch_size=64, shuffle=True)
train_loader_t = DataLoader(TensorDataset(X_t_tensor), batch_size=64, shuffle=True)

# Model
input_size = 128 * 3  # 128 time steps  3 features
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
encoder = FrequencyEncoder(input_size=input_size, hidden_size=128)
classifier = OpenSetClassifier(input_size=128, num_classes=len(known_classes))
model = RAINCOAT(encoder, classifier).to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.CrossEntropyLoss()

# Training
train_calot(model, train_loader_s, train_loader_t, optimizer, criterion, device)

# Evaluation
acc_known, acc_unknown, h_score = evaluate(model, X_t.numpy(), y_t.numpy(), device)
print(f"Known Accuracy: {acc_known:.4f}")
print(f"Unknown Accuracy: {acc_unknown:.4f}")
print(f"H-Score: {h_score:.4f}")

from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

def extract_features(model, loader_s, loader_t, device):
    model.eval()
    feats, labels, domains = [], [], []
    with torch.no_grad():
        # Source
        for x_s, y_s in loader_s:
            x_s = x_s.to(device).view(-1, 128, 3)
            _, _, f_s = model(x_s)
            feats.append(f_s.cpu())
            labels.append(y_s)
            domains.append(torch.zeros_like(y_s))  # 0 = source

        # Target
        for x_t, in loader_t:
            x_t = x_t.to(device).view(-1, 128, 3)
            _, _, f_t = model(x_t)
            feats.append(f_t.cpu())
            labels.append(torch.full((x_t.size(0),), -1))  # class unknown in target
            domains.append(torch.ones(x_t.size(0)))         # 1 = target

    feats = torch.cat(feats).numpy()
    labels = torch.cat(labels).numpy()
    domains = torch.cat(domains).numpy()

    return feats, labels, domains

def plot_tsne(feats_before, feats_after, labels, domains):
    from sklearn.manifold import TSNE
    import matplotlib.pyplot as plt
    import numpy as np

    # Compute t-SNE projections
    reduced_before = TSNE(n_components=2, perplexity=30, learning_rate='auto', init='random').fit_transform(feats_before)
    reduced_after  = TSNE(n_components=2, perplexity=30, learning_rate='auto', init='random').fit_transform(feats_after)

    fig, axs = plt.subplots(2, 2, figsize=(14, 10))

    # ---------- BY CLASS ---------- #
    for cls in np.unique(labels):
        mask = labels == cls
        label = f"Class {cls}" if cls != -1 else "Unknown"
        color = 'gray' if cls == -1 else None  # Let matplotlib assign color automatically
        axs[0, 0].scatter(reduced_before[mask, 0], reduced_before[mask, 1], label=label, alpha=0.6, s=10, c=color)
        axs[1, 0].scatter(reduced_after[mask, 0], reduced_after[mask, 1], label=label, alpha=0.6, s=10, c=color)

    axs[0, 0].set_title("Before Training - Colored by Class")
    axs[1, 0].set_title("After Training - Colored by Class")

    # ---------- BY DOMAIN (WITH SEPARATE TARGET-KNOWN / UNKNOWN) ---------- #
    # Masks
    source_mask        = (domains == 0) & (labels != -1)
    target_known_mask  = (domains == 1) & (labels != -1)
    target_unknown_mask= (domains == 1) & (labels == -1)

    # Colors
    source_color = "#1f77b4"   # Blue
    target_color = "#ff7f0e"   # Orange
    unknown_color = "#2ca02c"  # Green

    # Plot before training
    axs[0, 1].scatter(reduced_before[source_mask, 0], reduced_before[source_mask, 1], label="Source-known", c=source_color, alpha=0.6, s=10)
    axs[0, 1].scatter(reduced_before[target_known_mask, 0], reduced_before[target_known_mask, 1], label="Target-known", c=target_color, alpha=0.6, s=10)
    axs[0, 1].scatter(reduced_before[target_unknown_mask, 0], reduced_before[target_unknown_mask, 1], label="Target-unknown", c=unknown_color, alpha=0.6, s=10)

    # Plot after training
    axs[1, 1].scatter(reduced_after[source_mask, 0], reduced_after[source_mask, 1], label="Source-known", c=source_color, alpha=0.6, s=10)
    axs[1, 1].scatter(reduced_after[target_known_mask, 0], reduced_after[target_known_mask, 1], label="Target-known", c=target_color, alpha=0.6, s=10)
    axs[1, 1].scatter(reduced_after[target_unknown_mask, 0], reduced_after[target_unknown_mask, 1], label="Target-unknown", c=unknown_color, alpha=0.6, s=10)

    axs[0, 1].set_title("Before Training - Colored by Domain")
    axs[1, 1].set_title("After Training - Colored by Domain")

    # Legend and formatting
    for ax in axs.flat:
        ax.legend(markerscale=2)
        ax.set_xticks([])
        ax.set_yticks([])

    plt.tight_layout()
    plt.show()




# ----------------------
# Run t-SNE visualization
# ----------------------

# Extract features BEFORE training
encoder_eval = FrequencyEncoder(input_size=input_size, hidden_size=128)
classifier_eval = OpenSetClassifier(input_size=128, num_classes=len(known_classes))
model_eval = RAINCOAT(encoder_eval, classifier_eval).to(device)
model_eval.load_state_dict(model.state_dict())  # clone weights
feats_before, labels, domains = extract_features(model_eval, train_loader_s, train_loader_t, device)

# Train the model (already done earlier)

# Extract features AFTER training
feats_after, _, _ = extract_features(model, train_loader_s, train_loader_t, device)

# Plot both
plot_tsne(feats_before, feats_after, labels, domains)
```
